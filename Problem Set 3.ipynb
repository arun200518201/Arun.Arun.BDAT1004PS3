```python
# importing necessary libraries
import pandas as pd
```

```python
# importing dataset and assigning it to variable 'user'
users = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep = '|')
users.head()
```

```python
# mean age per population
users.groupby('occupation').age.mean()
```

```python
# male ratio per occupation in descending order
male_ratio_occupation = users["gender"].eq('M').groupby(users["occupation"]).mean().reset_index()
male_ratio_occupation.sort_values(by = ['gender'], ascending = False, ignore_index = True)
```

```python
# min and max age in each occupation
users.groupby(['occupation']).age.agg(['min','max'])
```

```python
# mean age for each combination of occupation and sex
users.groupby(["occupation",'gender']).age.mean()
```

```python
# percentage of men and women in each occupation
users.groupby(['occupation'])['gender'].value_counts(normalize=True)
```

## Question 4

```python
# importing necessary libraries
import pandas as pd
import datetime as datetime
```

```python
# importing dataset, assigning to variable data and replacing the first 3 columns by a proper datetime index
data = pd.read_csv('wind.txt', sep = '\s+', parse_dates = [[0,1,2]])
data.head()
```

```python
# Fixing year 2061
def fix_year(year):
    if year.year > 1989:
        fixed_year = year.year - 100
    else:
        fixed_year = year.year
        
    return datetime.date(fixed_year, year.month, year.day)

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix_year)
data.head()
```

```python
# Set the right dates as the index
data["Yr_Mo_Dy"] = pd.to_datetime(data["Yr_Mo_Dy"])
data = data.set_index('Yr_Mo_Dy')
data.index
```

```python
# number of missing for each location over the entire record
data.isnull().sum()
```

```python
# drop the missing values
data.dropna()
```

```python
# number of non-missing values
data.notnull().sum().sum()
```

```python
# mean windspeeds of the windspeeds over all the locations and all the times
data.sum().mean()/data.sum().sum()
```

```python
# DataFrame called loc_stats
# calculate the min, max and mean windspeeds and standard deviations of the windspeeds 
# at each location over all the days
loc_stats = data.describe()
loc_stats.loc[['min','max','mean', 'std']]
```

```python
# DataFrame called day_stats 
# and calculate the min, max and mean windspeed and standard deviations of the windspeeds 
# across all the locations at each day.

day_stats = pd.DataFrame()

day_stats['min'] = data.min(axis = 1)
day_stats['max'] = data.max(axis = 1) 
day_stats['mean'] = data.mean(axis = 1)
day_stats['std'] = data.std(axis = 1)

day_stats
```

```python
# average windspeed in January for each location
data.loc[data.index.month == 1].mean()
```

```python
# Downsample the record to a yearly frequency for each location.
data.groupby(data.index.to_period('A')).sum()
```

```python
# Downsample the record to a monthly frequency for each location.
data.groupby(data.index.to_period('M')).sum()
```

```python
# Downsample the record to a weekly frequency for each location.
data.groupby(data.index.to_period('W')).sum()
```

```python
# Calculate the min, max and mean windspeeds and standard deviations of the windspeeds 
# across all locations for each week for the first 52 weeks.
weekly_data = data.resample('W').agg(['min','max','mean','std'])
weekly_data.loc[weekly_data.index[:53]]
```

```python
# importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# importing data set
marriages_and_divorces = pd.read_csv('us-marriages-divorces-1867-2014.csv')
marriages_and_divorces.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(marriages_and_divorces['Year'], marriages_and_divorces['Marriages_per_1000'], label = 'Marriages', color = 'red')
plt.plot(marriages_and_divorces['Year'], marriages_and_divorces['Divorces_per_1000'], label = 'Divorces', color = 'green')
plt.title('Line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014')
plt.xlabel('Year')
plt.ylabel('Marriages/Divorces per capitaa')
plt.legend()
plt.show()
```

## Question 7

```python
years =[1900, 1950, 2000]
filtered_data = marriages_and_divorces[marriages_and_divorces.Year.isin(years)]
filtered_data.drop(['Marriages','Divorces', 'Population'], axis = 1, inplace = True)
filtered_data.set_index('Year', inplace = True)
filtered_data
```

```python
filtered_data.plot.bar()
plt.xlabel('Year')
plt.ylabel('Number')
plt.title('Number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000')
plt.show()
```

## Question 9

```python
roman_df = pd.read_csv('roman-emperor-reigns.csv')
roman_df.head()
```

```python
death_cause = roman_df.groupby(['Cause_of_Death']).count()[['Emperor']]
death_cause.head()
```

```python
cause = pd.Series(death_cause.index)
plt.figure(figsize=(12,10))
plt.pie(death_cause.Emperor, labels = cause, autopct='%.1f%%', explode =(0.1,0,0,0,0,0,0,0))
plt.title('Assassinated Roman Emperors')
plt.legend(title = 'Cause of Death', bbox_to_anchor=(0.5, -0.04), ncol=4)
fig = plt.gcf()
plt.show()
```
